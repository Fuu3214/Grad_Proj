```python
jacobian_transform = jacobian(model.hidden, model.x_input)

loss = tf.reduce_sum(tf.nn.sparse_softmax_cross_entropy_with_logits(labels=model.y_input, logits=model.pre_softmax))
grad_with_loss = tf.gradients(loss, model.x_input)[0]
```

```python
label_mask = tf.one_hot(model.y_input,
                              10,
                              on_value=1.0,
                              off_value=0.0,
                              dtype=tf.float32)
tar_logit = tf.reduce_sum(label_mask * model.pre_softmax, axis=1)
min_logit = tf.reduce_min((1-label_mask) * model.pre_softmax, axis=1)
cw_loss = -tf.nn.relu(tar_logit - min_logit + 50)
grad_with_cw_loss = tf.gradients(cw_loss, model.x_input)[0]
```

```python
def targeted_attack(nat, img, target_label, session, attack_epsilon, alpha = 1e-3):
    nn_res, logits, jacobian = session.run([model.y_pred, model.pre_softmax, jacobian_transform], feed_dict={model.x_input: img})
    jacobian = jacobian.reshape(-1,784)
    d_v_norm = np.delete(logits[0][target_label]-logits[0], target_label)
    v = np.delete(weights[:,target_label]-weights.transpose(), target_label, axis = 0)
    
    a = np.matmul(v, jacobian)
    epsilon = (np.abs(d_v_norm)/LA.norm(a, axis=1, ord=1))
    epsilon[d_v_norm>=0]=99999
    target = np.argmin(np.absolute(epsilon))
    if epsilon[target] == 99999:
        step_size = 0
    else:
        step_size = alpha
    eta = (step_size) * np.sign(a[target])
#     print("     choose decision_func:" + str(target))
#     print("epsilon: ")
#     print(epsilon)
#     print(epsilon[target])
#     print("d_v_norm: ")
#     print(d_v_norm)
#     print("logits: ")
#     print(logits)
#     print("--------------------")
    adv = np.clip(np.clip(img+eta, nat-attack_epsilon, nat+attack_epsilon), 0, 1)
    return nn_res[0], epsilon, d_v_norm, adv
#     return nn_res[0], epsilon, d_v_norm, img+eta
def i_dont_know_how_this_work_iter_attack_targeted(img, session, target, attack_epsilon, max_iter=50, no=1):
    nat = img
    img = img + np.random.uniform(-attack_epsilon, attack_epsilon, img.shape)
    cnt = 0;
    label = nn_classify(img, session)
    print("label: " + str(label))
#     for i in range(100):
#         cnt = cnt + 1
#         nn_res, epsilon, _, img = get_close_to_everything(nat, img, label, session)
        
#     im = img.reshape(28, 28)
#     im = PIL.Image.fromarray(im)
#     im.save('./non-targeted/non-targeted_adv%s.tiff' %no, 'tiff')
#     plt.imshow(img.reshape(28,28), cmap='Greys_r')
#     print("+++++++++++++++++++++++++++++++++++")
#     print("  adv_label: " + str(nn_res))
#     print("  iters: " + str(cnt))
#     print("+++++++++++++++++++++++++++++++++++")
            
    for i in range(max_iter):
        cnt = cnt + 1
        nn_res, epsilon, _, img = targeted_attack(nat, img, target, session, attack_epsilon)
        if(np.where(epsilon!=99999)[0].size == 0):
            break;
    
    if(nn_res == target):
        im = img.reshape(28, 28)
        im = PIL.Image.fromarray(im)
        im.save('./targeted/targeted_adv%s.tiff' %no, 'tiff')
        plt.imshow(img.reshape(28,28), cmap='Greys_r')
#         print("+++++++++++++++++++++++++++++++++++")
#         print("  adv_label: " + str(nn_res))
#         print("  iters: " + str(cnt))
#         print("+++++++++++++++++++++++++++++++++++")
#             print(get_majority_vote(distances, n_classes))
        return True, img;
            
    return False, img

def PGD(x_nat, y, sess, epsilon = 0.3, a = 1e-3, steps = 500):
    """Given a set of examples (x_nat, y), returns a set of adversarial
       examples within epsilon of x_nat in l_infinity norm."""
    x_adv = np.copy(x_nat) + np.random.uniform(-epsilon, epsilon, x_nat.shape)

    for i in range(steps):
        grad = sess.run(grad_with_cw_loss, feed_dict={model.x_input: x_nat,
                                            model.y_input: y})
        x_adv -= a * np.sign(grad)

        x_adv = np.clip(x_adv, x_nat - epsilon, x_nat + epsilon) 
        x_adv = np.clip(x_adv, 0, 1) # ensure valid pixel range

    return x_adv


def eval_example(adv, img):    
    print(np.amax(np.abs(img - adv)))
    print(np.where(adv>1)[0].size)
    print(np.where(adv<0)[0].size)
    print(np.amax(adv))
    print(np.amin(adv))
    
    print(nn_classify(adv,session))
    
```

```python
mnist = input_data.read_data_sets('MNIST_data', one_hot=False)
test_data = mnist.test.images
test_labels = mnist.test.labels

```

```python
from pgd_attack import LinfPGDAttack

attack = LinfPGDAttack(model, 
                       epsilon = 0.2,
                       k=500,
                       a=1e-3,
                       random_start=True,
                       loss_func = 'cw')


labeled_pred = model.softmax_layer[:,model.y_input[0]]
grad = tf.gradients(labeled_pred, model.x_input)
def integrated_gradient(img, target_label_index, steps = 50, baseline=None):
    if baseline is None:
        baseline = 0*img
    assert(baseline.shape == img.shape)
    steps=steps

    # Scale input and compute gradients.
    scaled_inputs = [baseline + (float(i)/steps)*(img-baseline) for i in range(0, steps+1)]

    gradient = session.run(grad, feed_dict = {model.x_input:np.squeeze(scaled_inputs),model.y_input:target_label_index})
    avg_grads = np.average(gradient[0][:-1], axis=0)
    integrated_gradients = (img-baseline)*avg_grads  # shape: <inp.shape>
    return integrated_gradients
```

```python
test_imgs = []
test_attr = []
adv_imgs = []
adv_attr = []
adv_labels = []
for i in range(200):
    img = np.expand_dims(test_data[i], axis=0)
    x_adv = attack.perturb(np.asarray(img), np.asarray([test_labels[i]]), session)
    y_adv = nn_classify(x_adv,session)
    if(test_labels[i] != y_adv):
        mask_test = integrated_gradient(test_data[i], [test_labels[i]])
        mask_adv = integrated_gradient(x_adv, [y_adv])
        
        test_imgs.append(img)
        adv_imgs.append(x_adv)
        adv_attr.append(mask_adv)
        test_attr.append(mask_test)
        adv_labels.append(y_adv)
    print(test_labels[i])
    print(nn_classify(x_adv,session))
```
